```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#x = number of unique DNA barcodes
#n = number of cells
#Since the number of viruses is infinite, we can assume that a cell "choosing" a barcode does not decrease the chance of another cell choosing the same barcode
#50 nt long but first 33 are unique. The number of unique barcodes is 4^33 = 7.3786976e+19
#each cell can choose 1 of x unique barcodes. The number of cells will be the number of simulations
#4^33 seems to be too large for numpy to handle, and I'm not aware of anything to get around this. Here, I will simulate smaller numbers.
```

```python
#for every simmulation, let's ensure that 95% of the cells are unique
#x = 970000
#n = 100000
#A = np.random.randint(1, x, n)
def cell_pop_unique(a,n):
    unique_array = list(set(a))
    ratio = len(unique_array)/n #number of unique cells/population n
    #print("ratio of cells: ", ratio)
    if ratio > 0.95:
        return True
    return False

#print(cell_pop_unique(A,n))
```

```python
#Simulate 10,000 times. Count the number of True from cell_pop_unique, then take it out of # of simulations.
def simulate(sim_num, x, n):
    num_true = 0
    for _ in range(sim_num):
        #generate array
        A = np.random.randint(1,x,n)
        if cell_pop_unique(A, n) == True:
            num_true = num_true + 1

    ratio = num_true/sim_num
    #print("ratio of simulations: ", ratio)
    if ratio > 0.95:
        return True
    return False

#test this
print(simulate(10000,500,10000)) #1000000 simulations, 500 barcodes, 10000 cells
```
